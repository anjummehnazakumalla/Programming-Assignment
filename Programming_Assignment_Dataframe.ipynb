```python
#Create Dataframe sat
sat_url = 'https://andybek.com/pandas-sat'

sat = pd.read_csv(sat_url)
sat
```

Q13: Starting with our main dataframe (*sat*), create a pivot table that summarizes the <ins>total</ins> student enrollment (Student Enrollment) across all 5 boroughs (*Borough*).

Which boroughs have the highest and lowest high school student bodies?

```python
#Q13
sat['Percent Tested'] = pd.to_numeric(sat['Percent Tested'].str.rstrip('%'))
enrollment = sat.pivot_table(values='Student Enrollment', index='Borough', aggfunc='sum')
print(enrollment)
borough_highest = enrollment.idxmax()
borough_lowest = enrollment.idxmin()
print("Borough having the highest School Student Body:", borough_highest)
print("Borough having the lowest School Student Body:", borough_lowest)
```

Q14: Modify the pivot table from the step above to also reflect the average student body size (*Student Enrollment*) across boroughs, in the same pivot table.

```python
#Q14

enrollment = sat.pivot_table(values='Student Enrollment', index='Borough', aggfunc=['sum', 'mean'])
enrollment.columns = ['Total Students Enrollment', 'Average Students Enrollment']
borough_highest_total = enrollment['Total Students Enrollment'].idxmax()
borough_lowest_total = enrollment['Total Students Enrollment'].idxmin()
print("Borough having the highest Total Student Body:", borough_highest_total)
print("Borough having the lowest Total Student Body:", borough_lowest_total)
enrollment
```

Q15: Create a pivot table of high schools from the Queens borough (*Borough*), containing the SAT section scores (*SAT Section*, *Score*) as columns and the school name (*School Name*) as index. Sort the table in descending order by math section scores.

```python
#Q15
Queens_borough = sat[(sat['Borough'] == 'Queens') & (sat['School Name'].notna())]
Queens_borough_p = Queens_borough.pivot_table(values='Score', index='School Name', columns='SAT Section')
Queens_high_schools_p = Queens_borough_p.sort_values(by='Math', ascending=False)
Queens_high_schools_p
```
